### Step 1: Create a map that displays a point for every city in the `city_data_df` DataFrame. The size of the point should be the humidity in each city.

```python
%%capture --no-display
# Configure the map plot
import folium
import random

# Create a map centered on a specific location
map_center = [latitude, longitude]  # Replace with the desired coordinates for the map center
map_city = folium.Map(location=map_center)

# Create a dictionary to store unique colors for each city
city_colors = {}

# Create a feature group to hold the markers
marker_group = folium.FeatureGroup(name='Cities')

# Iterate over each row in the DataFrame
for index, row in city_data_df.iterrows():
    city = row['City']
    latitude = row['Lat']
    longitude = row['Lng']
    humidity = row['Humidity']

    # Check if the city already has a color assigned, if not, assign a new color
    if city not in city_colors:
        # Generate a random color code for the city
        color = '#{0:06x}'.format(random.randint(0, 256**3-1))
        city_colors[city] = color

    # Create a circle marker for each city with size based on humidity and unique color
    marker = folium.CircleMarker(
        location=[latitude, longitude],
        radius=humidity / 10,  # Adjust the scaling factor as needed for appropriate marker size
        color='black',
        fill=True,
        fill_color=city_colors[city],
        fill_opacity=0.6,
        tooltip=f"City: {city}<br>Latitude: {latitude}<br>Longitude: {longitude}<br>Humidity: {humidity}"
    )

    # Add the marker to the feature group
    marker.add_to(marker_group)

# Add the feature group to the map
marker_group.add_to(map_city)

# Fit the map to display all the markers
map_city.fit_bounds(marker_group.get_bounds())

# Create the legend for each city
legend_html = '''
     <div style="position: absolute; 
                 bottom: 50px; left: 50px; width: 120px; height: auto; 
                 background-color: white;
                 border:2px solid grey; z-index:9999; font-size:14px;
                 ">&nbsp;<b>City Legend</b><br>'''
for city, color in city_colors.items():
    legend_html += f'<p style="margin: 5px"><i class="fa fa-circle" style="color:{color}"></i>&nbsp;{city}</p>'
legend_html += '</div>'

# Add the legend to the map
map_city.get_root().html.add_child(folium.Element(legend_html))

# Display the map
map_city

```

### Step 4: For each city, use the Geoapify API to find the first hotel located within 10,000 metres of your coordinates.

```python
# Set parameters to search for a hotel
radius = 10000
params = {
    "categories": "accommodation.hotel",
    "filter": f"circle:{latitude},{longitude},{radius}",
    "bias": f"proximity:{latitude},{longitude},
    "apiKey": geoapify_key,
    "limit": 1
}

# Print a message to follow up the hotel search
print("Starting hotel search")

# Iterate through the hotel_df DataFrame
for index, row in hotel_df.iterrows():
    # get latitude, longitude from the DataFrame
    latitude = row['Lat']
    longitude = row['Lng']
    # YOUR CODE HERE
    
    # Add filter and bias parameters with the current city's latitude and longitude to the params dictionary
    params["filter"] = f"circle:{latitude},{longitude},{radius}"
    params["bias"] = f"proximity:{latitude},{longitude}
    
    # Set base URL
    base_url = "https://api.geoapify.com/v2/places"


    # Make and API request using the params dictionaty
    name_address = requests.get(base_url,params)
    
    # Convert the API response to JSON format
    name_address = name_address.json()
    
    # Grab the first hotel from the results and store the name in the hotel_df DataFrame
    try:
        hotel_df.loc[index, "Hotel Name"] = name_address["features"][0]["properties"]["name"]
    except (KeyError, IndexError):
        # If no hotel is found, set the hotel name as "No hotel found".
        hotel_df.loc[index, "Hotel Name"] = "No hotel found"
        
    # Log the search results
    print(f"{hotel_df.loc[index, 'City']} - nearest hotel: {hotel_df.loc[index, 'Hotel Name']}")

# Display sample data
hotel_df
```
